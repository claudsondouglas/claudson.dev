# Calcular valor por vendedor

```python
import pandas as pd
```

```python
dados = {
    'Empresa': ['OMNIBEES SOLUCOES EM TECNOLOGIA E MARKETING HOTELEIRO LTDA', 'OUTRA EMPRESA'],
    'Promotor de Vendas': ['Promotor 1', 'Promotor 1, Promotor 2'],
    'Diárias': [10, 10]}
df = pd.DataFrame(dados)

#df = pd.read_excel('./Produtividade_Novembro_2024.xlsx', sheet_name='Produtividade')
df = df.loc[df['Empresa'] == 'OMNIBEES SOLUCOES EM TECNOLOGIA E MARKETING HOTELEIRO LTDA']
df['Promotor de Vendas'] = df['Promotor de Vendas'].fillna('vazio')
```

```python
def expandir_promotores(row):
    if pd.isna(row['Promotor de Vendas']):
        return pd.DataFrame()

    promotores = [p.strip() for p in row['Promotor de Vendas'].split(',')]

    if len(promotores) == 1:
        return pd.DataFrame({'Promotor de Vendas': promotores, 'individual': row['Diárias'], 'compartilhada': 0})
    else:
        diarias_divididas = row['Diárias'] / len(promotores)  # divido o valor pela quantidade de vendedores
        return pd.DataFrame({'Promotor de Vendas': promotores, 'individual': 0, 'compartilhada': diarias_divididas})
```

```python
dataframes = [expandir_promotores(row) for _, row in df.iterrows()]
df_expandido = pd.concat(dataframes, ignore_index=True)
```

```python
resultado = df_expandido.groupby('Promotor de Vendas').sum().reset_index()
resultado['total'] = resultado['individual'] + resultado['compartilhada'] #somar p ter o total em vendas
```
